### Stationnarité

```python
def adf_test(series, title=''):
    result = adfuller(series, autolag='AIC')
    ADF_stat = round(result[0],3)
    p_value = round(result[1],3)
    adf_disp = f'ADF Statistic for {title}: {ADF_stat}'
    if p_value < 0.05:
        p_disp = f"p-value ({p_value}) < 0.05 --> la série est stationnaire"
    else : p_disp=f"p-value ({p_value}) > 0.05 --> la série n'est pas stationnaire"
    print(f"{adf_disp}, {p_disp}")

for ticker in tickers:
    adf_test(close_prices[ticker], title=ticker)
```

### Normalité

```python
def shapiro_test(data, title=''):
    stat, p_value = shapiro(data.dropna())
    p_value = round(p_value,3)
    if p_value < 0.05:
        print(f'p-value for {title} ({p_value}) < 0.05 --> la normalité de la distribution est rejetée.')
    else :
        print(f'p-value for {title} ({p_value}) > 0.05 --> la normalité de la distribution est acceptée.')

for ticker in tickers:
    shapiro_test(close_prices[ticker], title=ticker)
```

### Tests de causalité (Granger)

Est-ce que les autres actions influencent le prix des actions Tesla ?

```python
# Fonction pour effectuer les tests de causalité de Granger pour chaque entreprise
def granger_test(data, target, maxlag):
    results = {}
    for col in data.columns:
        if col != target:
            print(f'Test de causalité de Granger entre {col} et {target}')
            result = grangercausalitytests(data[[target, col]], maxlag=maxlag, verbose=True)
            results[col] = result
    return results

# Effectuer les tests de causalité de Granger avec un maximum de 5 retards
maxlag = 5
results = granger_test(close_prices, 'TSLA', maxlag)
```

```python
# Interprétation des résultats
for col, result in results.items():
    for lag, test_result in result.items():
        f_test = test_result[0]['ssr_ftest']
        p_value = round(f_test[1],3)
        if p_value < 0.05:
            print(f'Les prix des actions de {col} causent le prix des actions de Tesla pour {lag} lag(s) {p_value}')
        #else:
            #print(f'Les prix des actions {col} ne causent pas les prix des actions de Tesla pour {lag} lag(s) {p_value}')
```

Les valeurs de p des tests F indiquent si une variable exogène (par exemple, les cours des actions de Facebook) Granger cause les cours des actions de Tesla. Si la valeur de p est inférieure à 0,05 pour un nombre donné de retards, cela signifie que les cours des actions de Facebook peuvent aider à prédire les cours des actions de Tesla, suggérant qu'il peut être utile d'inclure cette variable dans vos modèles prédictifs.
